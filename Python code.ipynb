```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("E:/assignment 3/data_random sample_half_cleaned.csv",index_col = [0])
print(len(data))
data.dtypes
```

```python
data.head()
```

```python
# 利用正则表达式，提取出房屋交易的年份
import re

trade_year = data.tradetime.apply(lambda datetime:re.search(r"[0-9]{4}",datetime).group(0))
print(trade_year)
```

```python
# 将房屋交易的年份作为新变量插入表格
data.insert(2,'trade_year',trade_year)
data.head()
```

```python
from numpy import int64

data['trade_year']=data['trade_year'].astype(int64)
data.dtypes
```

```python
trade_m = data.tradetime.apply(lambda datetime:re.search(r"(/[1-9]/)|(/1[0-2]/)",datetime).group(0))
print(trade_m)
data.insert(2,'trade_m',trade_m)
```

```python
# 利用正则表达式，提取出房屋交易的月份，并插入表格
trade_month = data.trade_m.apply(lambda datetime:re.search(r"(1[0-2])|([1-9])",datetime).group(0))
print(trade_month)
data.insert(2,'trade_month',trade_month)
data['trade_month']=data['trade_month'].astype(int64)
data.head()
```

```python
print(data.isnull().any())
```

```python
#删除变量dom，因为此变量不用于建模而且还有空值，防止后边删去含有空值的列时出错
del data['dom']
```

```python
import numpy as np

data = data.replace(to_replace='?',value=np.nan)
data = data.replace(to_replace=np.nan,value=None)
data = data.dropna(how='any') 
print(data.isnull().any())
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

column_names = ["lng", "lat", "trade_year", "trade_month", "constructiontime", "square", "livingroom", "drawingroom", "kitchen", "bathroom", "floor", "buildingtype", "renovationcondition", "buildingstructure", "ladderratio", "elevator", "fiveyearsproperty", "subway", "district", "communityaverage", "totalprice"]

# 分割训练数据和测试数据
# 随机采样25%作为测试 75%作为训练

x_train,x_test,y_train,y_test = train_test_split(data[column_names[0:20]],data[column_names[20]],test_size=0.25,random_state=42)

# 训练数据和测试数据进行标准化处理
ss_x = StandardScaler()
x_train = ss_x.fit_transform(x_train)
x_test = ss_x.transform(x_test)

ss_y = StandardScaler()
y_train = ss_y.fit_transform(y_train.array.reshape(-1, 1))
y_test = ss_y.transform(y_test.array.reshape(-1, 1))
```

```python
# 支持向量机模型进行学习和预测
from sklearn.svm import SVR

# 线性核函数配置支持向量机
linear_svr = SVR(kernel="linear")
# 训练
linear_svr.fit(x_train, y_train)
# 预测 保存预测结果
linear_svr_y_predict = linear_svr.predict(x_test)

# 多项式核函数配置支持向量机
poly_svr = SVR(kernel="poly")
# 训练
poly_svr.fit(x_train, y_train)
# 预测 保存预测结果
poly_svr_y_predict = linear_svr.predict(x_test)
print(poly_svr_y_predict)

# 模型评估
# 线性核函数 模型评估
print("线性核函数支持向量机的默认评估值为：", linear_svr.score(x_test, y_test))
print("线性核函数支持向量机的R_squared值为：", r2_score(y_test, linear_svr_y_predict))
print("线性核函数支持向量机的均方误差为:", mean_squared_error(ss_y.inverse_transform(y_test),
                                              ss_y.inverse_transform(linear_svr_y_predict)))
print("线性核函数支持向量机的平均绝对误差为:", mean_absolute_error(ss_y.inverse_transform(y_test),
                                                 ss_y.inverse_transform(linear_svr_y_predict)))
# 多项式核函数支持向量机模型评估
print("多项式核函数支持向量机的默认评估值为：", poly_svr.score(x_test, y_test))
print("多项式核函数支持向量机的R_squared值为：", r2_score(y_test, poly_svr_y_predict))
print("多项式核函数支持向量机的均方误差为:", mean_squared_error(ss_y.inverse_transform(y_test),
                                           ss_y.inverse_transform(poly_svr_y_predict)))
print("多项式核函数支持向量机的平均绝对误差为:", mean_absolute_error(ss_y.inverse_transform(y_test),
                                              ss_y.inverse_transform(poly_svr_y_predict)))
```

```python
# 三种集成回归模型进行训练和预测
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor

# 随机森林回归
rfr = RandomForestRegressor()
# 训练
rfr.fit(x_train, y_train)
# 预测 保存预测结果
rfr_y_predict = rfr.predict(x_test)

# 极端随机森林回归
etr = ExtraTreesRegressor()
# 训练
etr.fit(x_train, y_train)
# 预测 保存预测结果
etr_y_predict = rfr.predict(x_test)

# 梯度提升回归
gbr = GradientBoostingRegressor()
# 训练
gbr.fit(x_train, y_train)
# 预测 保存预测结果
gbr_y_predict = rfr.predict(x_test)

# 模型评估
# 随机森林回归模型评估
print("随机森林回归的默认评估值为：", rfr.score(x_test, y_test))
print("随机森林回归的R_squared值为：", r2_score(y_test, rfr_y_predict))
print("随机森林回归的均方误差为:", mean_squared_error(ss_y.inverse_transform(y_test),
                                          ss_y.inverse_transform(rfr_y_predict)))
print("随机森林回归的平均绝对误差为:", mean_absolute_error(ss_y.inverse_transform(y_test),
                                             ss_y.inverse_transform(rfr_y_predict)))

# 极端随机森林回归模型评估
print("极端随机树回归的默认评估值为：", etr.score(x_test, y_test))
print("极端随机树回归的R_squared值为：", r2_score(y_test, gbr_y_predict))
print("极端随机树回归的均方误差为:", mean_squared_error(ss_y.inverse_transform(y_test),
                                            ss_y.inverse_transform(gbr_y_predict)))
print("极端随机树回归的平均绝对误差为:", mean_absolute_error(ss_y.inverse_transform(y_test),
                                               ss_y.inverse_transform(gbr_y_predict)))

# 梯度提升回归模型评估
print("梯度提升决策树回归的默认评估值为：", gbr.score(x_test, y_test))
print("梯度提升决策树回归的R_squared值为：", r2_score(y_test, etr_y_predict))
print("梯度提升决策树回归的均方误差为:", mean_squared_error(ss_y.inverse_transform(y_test),
                                            ss_y.inverse_transform(etr_y_predict)))
print("梯度提升决策树回归的平均绝对误差为:", mean_absolute_error(ss_y.inverse_transform(y_test),
                                               ss_y.inverse_transform(etr_y_predict)))
```
